```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("iris.csv")
```

```python
sns.kdeplot(data[data['Species'] == 'Iris-setosa']['PetalLengthCm'], label='Setosa')
sns.kdeplot(data[data['Species'] == 'Iris-versicolor']['PetalLengthCm'], label='Versicolor')
sns.kdeplot(data[data['Species'] == 'Iris-virginica']['PetalLengthCm'], label='Virginica')
plt.title('Petal Length Distribution by Species')
plt.legend()
plt.show()
```

```python
sns.pairplot(data, hue='Species')
plt.title('Pair Plot for Iris Dataset')
plt.show()
```

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=data)
plt.title('Scatter Plot of Petal Length vs Petal Width by Species')
plt.show()
```

```python
sns.violinplot(x='Species', y='PetalWidthCm', data=data)
plt.title('Violin Plot for Petal Width by Species')
plt.show()
```

```python
species_counts = data['Species'].value_counts()
species_counts.plot(kind='bar', color='skyblue')
plt.title('Bar Plot of Species Counts')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='Species', y='SepalLengthCm', data=data)
plt.title('Box Plot for Sepal Length by Species')
plt.show()
```

```python
plt.scatter(data[data['Species'] == 'Iris-setosa']['SepalLengthCm'], data[data['Species'] == 'Iris-setosa']['SepalWidthCm'], label='Setosa')
plt.scatter(data[data['Species'] == 'Iris-versicolor']['SepalLengthCm'], data[data['Species'] == 'Iris-versicolor']['SepalWidthCm'], label='Versicolor')
plt.scatter(data[data['Species'] == 'Iris-virginica']['SepalLengthCm'], data[data['Species'] == 'Iris-virginica']['SepalWidthCm'], label='Virginica')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot of Sepal Length vs Sepal Width by Species')
plt.legend()
plt.show()
```

```python
plt.hist(data[data['Species'] == 'Iris-setosa']['SepalLengthCm'], alpha=0.5, label='Setosa')
plt.hist(data[data['Species'] == 'Iris-versicolor']['SepalLengthCm'], alpha=0.5, label='Versicolor')
plt.hist(data[data['Species'] == 'Iris-virginica']['SepalLengthCm'], alpha=0.5, label='Virginica')
plt.title('Histogram of Sepal Length by Species')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
for species in data['Species'].unique():
    sns.histplot(data[data['Species'] == species]['SepalLengthCm'], kde=True, label=species)
plt.legend()
plt.title('Sepal Length Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.swarmplot(x='Species', y='SepalWidthCm', data=data)
plt.title('Sepal Width by Species')
plt.show()
```
